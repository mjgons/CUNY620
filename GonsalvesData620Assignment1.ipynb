```python
import networkx as nx
```

```python
import matplotlib.pyplot as plt
```

```python
# Create an empty graph
```

```python
G = nx.Graph()
```

```python
# Define the notes for the Graph
```

```python
nodes = list(range(10)) # Nodes labeled 0-9
```

```python
G.add_nodes_from(nodes)
```

```python
# Define the edges (connections between nodes)
```

```python
edges = [(0,1), (0,2), (0,3), (1,3), (2,3), (3,4), (4,5), 
         (5,6), (5,7), (6,7), (7,8), (8,9)]
```

```python
G.add_edges_from(edges)
```

```python
# Draw the Graph
```

```python
plt.figure(figsize=(6,6))
```

```python
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='grey', node_size=800, font_size=12)
```
